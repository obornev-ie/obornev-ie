```python
import matplotlib.pyplot as plt
from matplotlib.patches import ConnectionPatch
%config InlineBackend.figure_formats = ['svg']
def quadroCircle(xC,yC,R,level,direction=0):

    if (level<1) or (R==0): return
    if level>=2:
        plt.gca().add_patch(ConnectionPatch((xC+R*2,yC), (xC-R*2,yC),'data',color='c'))
        plt.gca().add_patch(ConnectionPatch((xC,yC+R*2), (xC,yC-R*2),'data',color='c'))
    
    plt.gca().add_patch(plt.Circle((xC,yC),R,color='c',fill=True))
    
    nextR=R/2
    l=[level for i in range(4)]
    if direction!=0: l[direction-1]=0
    quadroCircle(xC-R*2,yC,nextR,l[1]-1,1)
    quadroCircle(xC+R*2,yC,nextR,l[0]-1,2)
    quadroCircle(xC,yC-R*2,nextR,l[3]-1,3)
    quadroCircle(xC,yC+R*2,nextR,l[2]-1,4)
     
def quadroCirclePaint(xC,yC,R,level,direction=0):
    quadroCircle(xC,yC,R,level)
    plt.axis('scaled')
    plt.show()
quadroCirclePaint(0,0,10,5)
```
